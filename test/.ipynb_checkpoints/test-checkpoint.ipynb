# Подготовка данных

```python
import pandas as pd
import numpy as np
```

```python
train_raw = pd.read_csv("Learn_sample.txt", delimiter="\t", header=None)
test_raw = pd.read_csv("Test_sample.txt", delimiter="\t", header=None)
```

```python
train_raw.head()
```

```python
test_raw.head()
```

Удаляем лишние столбцы и инвертируем матрицы, чтобы привести в удобный для ML алгоритмов формат

```python
def prepare_data(data):
    result = data.copy()
    result = result.drop([0,1,2,3], axis=1)
    result = np.transpose(result)
    result = result.reset_index(drop=True)
    return result
```

```python
train = prepare_data(train_raw)
test = prepare_data(test_raw)
```

```python
train
```

```python
test
```

# Преобразуем маркеры в значения

```python
from sklearn.preprocessing import StandardScaler
```

Не смог найти в интернете примеров функций преобразования последовательностей нуклеотидов в значения для маркеров. Поэтому применил такую тревиальную. Она не отображает корректное растояние, но за счет большого количества маркеров и нормировки, это должно сгладится

```python
def value_metric1(marker):
    result = 0
    for n in marker:
        result *= 4
        if n == 'C':
            result += 1
        if n == 'G':
            result += 2
        if n == 'T':
            result += 3
    return result
```

```python
train_value1 = train.applymap(value_metric1)
```

```python
train_value1
```

```python
test_value1 = test.applymap(value_metric1)
```

```python
test_value1
```

Нормируем полученные фичи, чтобы небыло перевеса в сторону некоторых фич из-за не идеальности функции преобразования

```python
scaler = StandardScaler()
scaler.fit(train_value1)
train_value1_scaled = pd.DataFrame(scaler.transform(train_value1))
test_value1_scaled = pd.DataFrame(scaler.transform(test_value1))
```

# Иерархическая кластеризация

На сколько я понимаю для таких задач как правило используется иерархический алгоритм, хотя упоминания применения kmeans тоже есть

```python
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering
```

Сначала обучим модель без сегментов, чтобы нарисовать красивый график и посмотреть, что получается

```python
test_model1 = AgglomerativeClustering(distance_threshold=0, n_clusters=None)
test_model1 = test_model1.fit(train_value1_scaled)
```

```python
def plot_dendrogram(model, **kwargs):
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    dendrogram(linkage_matrix, **kwargs)
```

```python
plt.figure(figsize=(15,5))
plot_dendrogram(test_model1, truncate_mode='level', p=5)
plt.show()
```

График получился очень ровный, даже слишком ровный, с двумя ярко выраженными кластерами. Теперь обучим модель с двумя кластерами

```python
model1 = AgglomerativeClustering(n_clusters=2)
train_label1 = model1.fit_predict(train_value1_scaled)
```

# Классификация тестовой выборки

К сожалению для иерархического алгоритма в sklearn не реализован метод predict, поэтому придется в ручную сравнивать растояния каждой записи из тетовой выборки со средними значениями для обоих кластеров. Я буду брать манхетенское расстояние, т. к. оно как правило лучше всего работает для задачи иерархической кластеризации

```python
mean0 = train_value1_scaled.loc[train_label1[train_value1_scaled.index] == 0].mean(axis = 0)
mean1 = train_value1_scaled.loc[train_label1[train_value1_scaled.index] == 1].mean(axis = 0)
```

```python
mean0
```

```python
mean1
```

### Подозрительные средние значения, отличаются только знаком :) !!!!

```python
def calc_len( A, B):
    result = 0
    for a, b in zip(A, B):
        result += abs(a-b)
    return result
```

```python
test_label = []
for index, row in test_value1_scaled.iterrows():
    if calc_len(mean0, row) < calc_len(mean1, row):
        test_label.append(0)
    else:
        test_label.append(1)
test_label = np.array(test_label)
```
